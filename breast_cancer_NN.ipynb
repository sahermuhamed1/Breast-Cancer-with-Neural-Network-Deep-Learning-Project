```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_dataset = sklearn.datasets.load_breast_cancer()
breast_dataset
```

```python
df = pd.DataFrame(breast_dataset.data, columns = breast_dataset.feature_names)
df['label'] = breast_dataset.target
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['label'].value_counts()
```

```python
df.groupby('label').mean()
```

```python
x = df.drop(columns = 'label', axis = 1)
y = df['label']
x_train, x_test, y_train, y_test = train_test_split(x, y ,test_size = 0.2 , random_state=2)
print(x.shape,x_train.shape , x_test.shape)  
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(x_train)
X_test_scaled = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow import keras
tf.random.set_seed(3) 
    
model = keras.Sequential([
        keras.layers.Flatten(input_shape=(30,)),
        keras.layers.Dense(20, activation='relu'),
        keras.layers.Dense(2, activation='sigmoid')
    ])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
history = model.fit(X_train_scaled, y_train, validation_split=0.1, epochs=100)

```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt. legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt. legend(['training data', 'validation data'], loc = 'lower right')
```

```python
loss, accuracy = model.evaluate(X_test_scaled, y_test)
```
